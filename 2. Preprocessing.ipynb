```python
import re
ulasan = df

def cleaningUlasan(ulasan):
  ulasan = re.sub(r'@[A-Za-a0-9]+',' ',ulasan)
  ulasan = re.sub(r'#[A-Za-z0-9]+',' ',ulasan)
  ulasan = re.sub(r"http\S+",' ',ulasan)
  ulasan = re.sub(r'[0-9]+',' ',ulasan)
  ulasan = re.sub(r"[-()\"#/@;:<>{}'+=~|.!?,_]", ' ', ulasan)
  ulasan = re.sub(r"(?:\@|https?\://)\S+", ' ', ulasan)
  ulasan = re.sub(r'[^\x00-\x7f]', ' ', ulasan)
  ulasan = ulasan.strip(' ')
  return ulasan
ulasan['Cleaning'] = ulasan['text'].apply(cleaningUlasan)

def clearEmoji(ulasan):
  return ulasan.encode('ascii', 'ignore').decode('ascii')
ulasan['HapusEmoji'] = ulasan['Cleaning'].apply(clearEmoji)

def replaceTOM(ulasan):
  pola = re.compile(r'(.)\1{2,}', re.DOTALL)
  return pola.sub(r'\1', ulasan)
ulasan['3/Lebih']= ulasan['HapusEmoji'].apply(replaceTOM)

def casefoldingText(ulasan):
  ulasan = ulasan.lower()
  return ulasan
ulasan['CaseFolding']= ulasan['3/Lebih'].apply(casefoldingText)
```

```python
from nltk.corpus import stopwords

daftar_stopword = stopwords.words('indonesian')

daftar_stopword = set(daftar_stopword)

def stopwordText(words):
    return [word for word in words if word not in daftar_stopword]

ulasan['Stopword Removal'] = ulasan['Formalisasi'].apply(stopwordText)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in ulasan['Stopword Removal']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)

def stemmingText(document):
    return [term_dict[term] for term in document]

ulasan['Stemming'] = ulasan['Stopword Removal'].swifter.apply(stemmingText)
```

```python
def fit_stopwords(document):
    document = np.array(document)
    document = ' '.join(document)
    return document
df['Clean']=df['Stemming'].apply(lambda x: fit_stopwords(x))
```

```python
import csv

lexicon_positif = {}
with open('/content/drive/MyDrive/NLP/TUBES/data/positive.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        if row[1].isdigit():
            lexicon_positif[row[0]] = int(row[1])
        else:
            print(f"Nilai tidak valid untuk kunci {row[0]}: {row[1]}")

lexicon_negatif = {}
with open('/content/drive/MyDrive/NLP/TUBES/data/negative.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        try:
            lexicon_negatif[row[0]] = int(row[1])
        except ValueError:
            print(f"Nilai tidak valid untuk kunci {row[0]}: {row[1]}")
```

```python
def sentimen(text):
  score = 0
  for word in text:
    if (word in lexicon_positif):
      score = score + lexicon_positif[word]
  for word in text:
    if (word in lexicon_negatif):
      score = score + lexicon_negatif[word]

  polarity = ''
  if (score > 0):
    polarity = "Positive"
  elif (score < 0):
    polarity = "Negative"
  else:
    polarity = "Neutral"

  return score, polarity
```

```python
results = df['Stemming'].apply(sentimen)
results = list(zip(*results))
df['Score'] = results[0]
df['Label'] = results[1]
print(df['Label'].value_counts()), df.shape
```
